## Neue Spalten erstellen

```python
movies.loc[:, ['Rotten_Tomatoes_Rating', 'IMDB_Rating']]
```

```python
movies['Rotten_Tomatoes_Rating_10'] = movies['Rotten_Tomatoes_Rating'] / 10
```

```python
movies.Rotten_Tomatoes_Rating_10
```

```python
movies['Profit'] = movies['Worldwide_Gross'] - movies['Production_Budget']
```

```python
movies['ROI'] = movies['Profit'] / movies['Production_Budget']
```

```python
movies.sort_values('ROI', ascending=False)
```
